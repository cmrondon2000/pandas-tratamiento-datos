#**PROBLEMA DE NEGOCIO**
Se trata de um algoritmo capaz de analizar las características de una propiedad como comodidades, tamaño, ocupación del espacio en un período determiando y sugerir al anfitrión un precio a cobrar por tarifas diarias que garantice ganancia en momentos de alta demanda.

##** 1. Entendiendo el Problema

```python
import pandas as pd
```

```python
datos = pd.read_json("datos_hosting.json")
datos
```

```python
datos_n =pd.json_normalize(datos["info_inmuebles"])
datos_n.head()
```

```python
datos_n.info()
```

##** Datos Numericos**

```python
columnas = list(datos_n.columns)
columnas
```

```python
datos_ex = datos_n.explode(columnas[3:])
datos_ex
```

```python
datos_ex.reset_index(inplace=True, drop = True)
datos_ex.head()
```

```python
datos_ex.info()
```

```python
import numpy as np
```

```python
datos_ex["max_hospedes"] = datos_ex["max_hospedes"].astype(np.int64)
```

```python
datos_ex.info()
```

```python
col_numericas =["cantidad_baños", "cantidad_cuartos","cantidad_camas"]
```

```python
datos_ex[col_numericas] = datos_ex[col_numericas].astype(np.int64)
```

```python
datos_ex.info()
```

```python
datos_ex["evaluacion_general"] = datos_ex["evaluacion_general"].astype(np.float64)
```

```python
datos_ex.info()
```

```python
datos_ex["precio"]=datos_ex["precio"].apply(lambda x: x.replace("$","").replace(",","").strip())
```

```python
datos_ex.info()
```

```python
datos_ex["precio"] = datos_ex["precio"].astype(np.float64)
```

```python
datos_ex.info()
```

```python
datos_ex[["cuota_deposito", "cuota_limpieza"]] = datos_ex[["cuota_deposito", "cuota_limpieza"]].applymap(lambda x: x.replace("$","").replace(",","").strip())
datos_ex[["cuota_deposito", "cuota_limpieza"]] = datos_ex[["cuota_deposito", "cuota_limpieza"]].astype(np.float64)
```

```python
datos_ex.info()
```

```python
datos_ex["descripcion_local"]= datos_ex["descripcion_local"].str.lower()
```

```python
datos_ex["descripcion_local"][3169]
```

```python
datos_ex["descripcion_local"] = datos_ex["descripcion_local"].str.replace("[^a-zA-Z0-9\-\' ]","", regex=True)
```

```python
datos_ex["descripcion_local"] = datos_ex["descripcion_local"].str.replace("(?<!\w)-(?<!\w)"," ", regex=True)
```
